```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
```

```python
class Guide:
    def __init__(self,title,descURL,categories,thumb_url):
        #Set from baseCrawl
        self.title = title
        self.descURL = descURL 
        self.categories = categories
        self.thumb_url = thumb_url

        self.img_DigitalID = ""
        self.img_Description = ""
        self.img_URL = ""
        self.description_Short = ""
        self.description_CollectionHistory = ""
        self.description_Background = ""
        self.description_RelatedResources = ""
        #Mandatory
        self.description_Divisions = []

        #Mandatory
        #Collection Contents is hierarchical
        self.collectionContents = []
        self.collectionContents_link = ""

        #Optional
        self.relatedSubjects = []
        self.relatedSubjects_link = ""

        
        # And now we fetch every guide's major description page
        self.descURLr = requests.get(self.descURL)
        self.descSoup = BeautifulSoup(self.descURLr.text)
        time.sleep(.5)
    
    def __repr__(self):
        return "<Collection Guide: %s>" %self.title

        
    
```

```python
startURL = "http://digitalgallery.nypl.org/nypldigital/explore/dgexplore.cfm?topic=all"

r = requests.get(startURL)
soup = BeautifulSoup(r.text)

#Get the section of the page that lists all guides
guidetoguides = soup.find(id="featuredcontent")

#then split that section so it's iterable
guidelist = guidetoguides.findAll('span')

allGuides = []
#print("Number of guides: %d"%len(allGuides))

for gdiv in guidelist:
    
    gtitle = ""
    gdescURL = ""
    gcategories = []
    gthumb_url = []
    # Find & set guide title
    gtitle=gdiv.find('h4').getText()
    
    links = gdiv.find_all('a')
    #print(len(links))
    for l in links:
        #Either it's a master link / title, OR its a category tag
        if l.find('h4'):
            gdescURL = urljoin(r.url, l.get('href'))
        else:
            #print(l.string)
            gcategories.append(l.string)
    
    gthumb_url = gdiv.find('img').get('src')
    allGuides.append(Guide(gtitle,gdescURL,gcategories,gthumb_url))
    
    
```

```python
#Now it's time to parse the guide page

linkTypes = {}

for aguide in allGuides:
    
    #Div ID 'featuredcollection7' has the main navigation links
    fc7 = aguide.descSoup.find(id='featuredcollection7')
    
    for link in fc7.find_all('a'):
        if link.string in linkTypes:
            linkTypes[link.string] += 1
        else:
            linkTypes[link.string] = 1
    
    for link in fc7.find_all('a'):
        if link.string == "Collection Contents":
            print(urljoin(aguide.descURLr.url,link.get('href')))

#print(linkTypes)
```
